```python
yari_cap = 1.5
pi_degeri = 3.14

daire_alan = pi_degeri * (yari_cap*yari_cap)
daire_cevre = 2 * pi_degeri * yari_cap

print("Dairenizin Alanı: ", daire_alan)
print("Dairenizin Çevresi: ", daire_cevre)
```

```python
while True:
    try:
        sayi = int(input("Lütfen 4 basamaklı bir sayı girin: "))
        if 1000 <= sayi <= 9999:
            break
        else:
            print("Girdiğiniz sayı 4 basamaklı değil. Tekrardan sayı giriniz.")
    except ValueError:
        print("Geçersiz giriş. Bir sayı girmelisiniz.")

ilk_basamak = sayi // 1000
son_basamak = sayi % 10
toplam = ilk_basamak + son_basamak

print(f"Girdiğiniz sayının ilk ve son basamağı toplamı: {toplam}")
```
