```python
x = int(input("Lutfen bir sayı giriniz : "))
if (x % 7 == 0 or x % 5 == 0) :
  if (x % 7 == 0 and x % 5 == 0) :
    print("Girmiş olduğunuz sayı hem 5'e hem de 7'ye tam bölünmektedir")
  elif (x % 7 == 0) :
    print("Girmiş olduğunuz sayı 7'ye tam bölünmektedir")
  else :
    print("Girmiş olduğunuz sayı 5'e tam bölünmektedir")
else :
  print("Girdiğiniz sayı ne 5'e ne de 7'ye tam bölünmektedir")
```

```python
kilo = float(input("Lütfen kilonuzu giriniz (kg) : "))
boy = int(input("Lütfen boyunuzu giriniz (cm) : "))
boy = boy / 100
BMI = kilo / (boy ** 2)
if (BMI >= 30) :
  print ("OBEZSİN!!! Kilo vermelisin")
elif (BMI >= 25 and BMI < 30) :
  print ("Aşırı kilolusun, obez değilim diyip kendini kandırmayı kes. Hemen kilo ver")
elif (BMI >= 20 and BMI < 25) :
  print ("Sen normalsin, aynen yola devam")
else :
  print ("Fazla zayıfsın. Ölmeden önce hala yemek yemen için ömnünde bir fırsat var")
```
