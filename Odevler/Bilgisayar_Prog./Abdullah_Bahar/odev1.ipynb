```python
K = 400
sabit = 273.15
C = K - sabit
print(f'{K} Kelvin = {C} Celcius')
```

```python
r = 1.5
pi = 3.14
cevre = 2 * r * pi
alan = pi * r ** 2
print(f"{r} cm'lik bir yari capa sahip dairenin alani = {alan} ve cevresi = {cevre}" )
```

```python
# bu kod 4 basamakli bir sayi disindeki sayilarda da calisacaktir
x = input("Lutfen 4 basamakli bir sayi giriniz : ")
ilk = int(x[0])
son = int(x[-1])
toplam = ilk + son
print(f'Girmis oldugunuz sayilarin ilk ve son basamaklarinin toplami :')
print(f'{ilk} + {son} = {toplam}')
```

```python
hiz = int(input("Lutfen hiz degerini giriniz (km/sa) : "))
yol = int(input("Lutfen katedilen mesafeyi giriniz (km) : "))
time = yol / hiz
saat = int(time)            # zamanin tam sayi kismini aldım
dakika = time - int(time)   # zamanin ondalikli kismini aldim
dakika = round(dakika * 60) # ondalikli kisim saat cinsinde deger gosterecegi icin dakikaya çevirdik. Ayrıca küsürattan da kurtulduk.
print(f"{hiz} km/sa hıza sahip bir arac, {yol} km yolu yaklasik olarak {saat} saat, {dakika} dakikada tamamlar.")
```
